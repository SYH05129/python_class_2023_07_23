```python
x=np.linspace(-5,5,num=20)
y1=5*x+1
y2=x**3
import matplotlib.pyplot as plt
figure=plt.figure(figsize=(8,5))
axes=figure.add_subplot(1,1,1)
axes.plot(x,y1,"b--")
axes.plot(x,y2,"g--")
plt.show()
```

```python
t = np.arange(0,5,0.1)
y1 = np.sin(2 * np.pi * t)
y2 = np.cos(2 * np.pi * t)

figure = plt.figure(figsize=(8,5))
axes1 = figure.add_subplot(2,1,1)
axes2 = figure.add_subplot(2,1,2)
axes1.plot(t,y1,'r--')
axes2.plot(t,y2,'b--')
plt.show()
```
