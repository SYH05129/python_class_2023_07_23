```python
import matplotlib.pyplot as plt
import numpy as np
index=np.arange(5)
values=np.array([5,7,3,4,6])*1000
figure=plt.figure(figsize=(8,5))
axes=figure.add_subplot(1,1,1)
axes.bar(index,values)
axes.set_xticks(index,['apple', 'banana', 'orange', 'totato', 'guava'])
axes.grid(axis='y')
xmin,xmax,ymin,ymax=axes.axis()
axes.axis([xmin-0.1, xmax+0.1, ymin, ymax+1000])
for i in index:
    print(i,values[i],end=' ')
    axes.annotate(str(values[i]),[i-0.15,values[i]+200])
   
plt.show()
```
