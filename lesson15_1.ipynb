```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font=FontProperties(fname=r'/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)

student1={"國文":78,"英文":80,"數學":92,"歷史":75,"地理":85}
student2={"國文":90,"英文":75,"數學":98,"歷史":65,"地理":75}
figure=plt.figure(figsize=(8,5),dpi=250,facecolor="#bbbb")
axes=figure.add_subplot(1,1,1)
axes.plot(student1.keys(),student1.values(),"g--o",label='stu1')
axes.plot(student2.keys(),student2.values(),"r-+",label='stu2')
axes.set_xticklabels(student1.keys(),fontproperties=font)
axes.legend()
axes.set_title("學生1和學生2分數比較表",fontproperties=font)
axes.set_xlabel("科目",fontproperties=font)
axes.set_ylabel("分數",fontproperties=font)

for x in axes.get_xticks():
    value1=list(student1.values())[x]
    value2=list(student2.values())[x]
    axes.annotate(str(value1),(x,value1+0.5),fontproperties=font,color="green")
    axes.annotate(str(value2),(x,value2+0.5),fontproperties=font,color="red")
```

```python
import random
import matplotlib.pyplot as plt
days=list(range(1,31))
celsius=[round(random.uniform(24.5,37.5),ndigits=1) for _ in range(1,31)]
figure=plt.figure(figsize=(8,5),dpi=250)
axes=figure.add_subplot(1,1,1)
axes.plot(days,celsius,"r-+",label="September")
axes.grid()
axes.legend()
axes.set_title("Weather")
axes.set_xlabel("Day")
axes.set_ylabel("Celsius")
axes.set_xticks(days)
plt.show()
```
