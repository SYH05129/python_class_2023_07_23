```python
import matplotlib.pyplot as plt

fig=plt.figure()

ax_1=fig.add_subplot(1,1,1)

x=[0,1,2]
y=[3,4,5]
ax_1.plot(x,y)

plt.show()
```

```python
figure=plt.figure()
ax1=figure.add_subplot(1,3,1)
ax1.set_title("Axes1")
ax2=figure.add_subplot(1,3,2)
ax2.set_title("Axes2")
ax3=figure.add_subplot(1,3,3)
ax3.set_title("Axes3")
```

```python
figure=plt.figure(figsize=(8,5),dpi=220)
stu1={"Chinese":80,"math":95,"science":75,"histery":88,"English":90}
stu2={"Chinese":90,"math":85,"science":70,"histery":77,"English":87}
axes=figure.add_subplot()
axes.plot(stu1.keys(),stu1.values())
axes.plot(stu2.keys(),stu2.values())
plt.show()
```
